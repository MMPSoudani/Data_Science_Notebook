## 1.1. What is Machine Learning
<div style="font-weight: bold; text-align: justify; font-size: 15px">
Machines perform tasks given by the instructor. These tasks needs to be described in a way that the machine can understand them. Oftentimes, the instructor needs to describe the steps the machine needs to take in order for the machine to perform that task. In this case, machine learning is a process that allows the machine to learn and understand how to perform its given tasks and also how to use the gained knowledge to perform more complex tasks.
</div><br>

<div style="font-weight: bold; text-align: justify; font-size: 15px">
The main difference between a computer algorithm and a machine learning algorithm, is the learning process. The computer algorithm takes the input and the instructions then creates an output. the machine learning algoritm takes the input and the output and tries to learn the instructions for producing the output.
</div>

## 2.2. Types of machine learning

<div style="font-weight: bold; text-align: justify; font-size: 15px">
Each of the following types uses different algorithms, but the main goal of all of them is to help the machine learn based on the input data and make a prediction on the outcome of simillar data.
</div>

### 2.2.1. Supervised
<div style="font-weight: bold; text-align: justify; font-size: 15px">
In this ML type, the input data is already categorized or have labels. In the supervised learning method, the input data can be Classified into categorize, thus the model will decide the new data belongs to which group or category, or use Regression, which aims to model the relationship between a certain number of features and a continuoues target value.
</div>

### 2.2.2. Unsupervised
<div style="font-weight: bold; text-align: justify; font-size: 15px">
In this ML type, the input data does not have a category or labels. In the unsupervised learning method, the input data can grouped by the machine using Clustering algorithms or by using Association Rule Learning which aims to discovere interesting relations between variables in large databases. It is intended to identify strong rules discovered in databases using some measures of interestingness.
</div>

### 2.2.3. Transfer learning
<div style="font-weight: bold; text-align: justify; font-size: 15px">
In this ML type, the results of another model can be used instead of recalculating the same model for the same input data to save time and energy.
</div>

### 2.2.4. Reinforcment
<div style="font-weight: bold; text-align: justify; font-size: 15px">
In this ML type, the machine learns to complete its tasks by trial and error or by rewards and punishments.
</div>

## 2.4. Evaluation Metrics

<div style="font-weight: bold; text-align: justify; font-size: 15px">
Evaluation metrics helps decide how well the ML model performs and measure the ML model success.
Evaluation metrics to consider for Classification problems:<br>
    <ol>
        <li>False Negatives: the models predicts its a negative, but its actually a positive.</li>
        <li>False Positives: the models predicts its a positive, but its actually a negative.</li>
        <li>True Negatives: the models predicts its a negative and its correct.</li>
        <li>True Positives: the models predicts its a positive and its correct.</li>
        <li>Precision: what proportion of the positive predections are actually correct?</li>
        <li>Recall: what proportion of the actual positive were predicted correctly?</li>
        <li>F1 Score: a combination of precision and recall</li>
        <li>Receiver Operating Characteristics (ROC) Curve: a plot that compares the True Positives and False Positives.</li>
        <li>Area Under the Curve (AUC): the area under the ROC curve.</li>
    </ol><br>

Evaluation metrics to consider for Regression problems:
    <ol>
        <li>Mean Absolute Error (MAE): The average difference between the model's preditions and the actual numbers.</li>
        <li>Root Mean Squre Error (RMSE): The square root of the average of squared differences between your model's predictions and the actual numbers.</li>
    </ol>
</div>

## 2.5. Features in Data

<div style="font-weight: bold; text-align: justify; font-size: 15px">
Features refers to other forms of data. and means what do we already know about the data?
Feature Variables are what the model uses in order to predict the Traget Variables.
The main three types of features are:<br>
    <ol>
        <li>Categorical Features: Yes/No, Male/Female</li>
        <li>Continuous or Numerical Features</li>
        <li>Derived Features: features created from the data. referred to as feature engineering.</li>
    </ol>
<span style="font-weight: 900">Feature Engineering:</span> "how a subject matter expert takes their knowledge and encode it into the data".
importing note to consider when it comes to features:<br>
    <ol>
        <li>A machine learning model should be trained on features which represent as close as possible to what it will be used
    for in a real system.
        <li>Let your machine learning engineers and data scientists know: What do you already know about the problem and how can that influence what features you use?.
        <li>Have a preference for features with the most coverage. The ones where lots of samples have data for.
        <li>If your model is achieving perfect performance, youâ€™ve likely got feature leakage somewhere. Which means the data your model has trained on is being used to test it. No model is perfect.
    </ol>
</div>

## 2.6. Modeling

<div style="font-weight: bold; text-align: justify; font-size: 15px">
After defining the problem, preparing the data, evaluation criteria and features, its time to model.<br>

The most important concept in machine learning: The 3 Sets which means splitting the input data into a training set, a validation set and a testing set. Usually the data splitt happens at 70% for training, 15% for validation and 15% for testing.
Generalization is the ability for a machine learning model to perform well on data it has not seen before. This can happen due to what the model has learned from on a different dataset.<br>

Modeling can be broken down into three parts:<br>
    <ol>
        <li>Choosing and Training a Model</li>
        <li>Tuning and Improving a Model</li>
        <li>Comparing a Model</li>
    </ol>

Notes to remember:
    <ol>
        <li>Some models may work better than other models on different problems</li>
        <li>Try many different ways</li>
        <li>Start Small and build up (add complexity) as you need</li>
        <li>Machine learning models have hyper-parameters than can be adjusted during the training and validation phase</li>
        <li>When comparing the models, choose the most optimal model based on its training time, prediction time and performance</li>
    </ol>

Overfitting and Underfitting means the ML model is not able to generalize well
    <ol>
        <li>Overfitting: The test performance is higher than the train performance. one reason can be test data leakage into training data</li>
        <li>Underfitting: The test performance is lower than the train performance. one reason can be data mismatch between training dataset and test dataset.</li>
    </ol>

To fix Overfitting try:
    <ol>
        <li>Collect more data</li>
        <li>Try a less advanced ML model</li>
    </ol>
To fix Underfitting try:
    <ol>
        <li>Try a more advanced model</li>
        <li>Increase models hyper-parameters</li>
        <li>Reduce amount of features</li>
        <li>Train longer</li>
    </ol>
</div>

## 5.6. Mathematical Operations on Arrays
<div style="font-weight: bold; text-align: justify; font-size: 15px">
    All mathematical operations like +, -, *, /, //, %, ** can be applied to NumPy arrays.<br>
    When applying these mathematical operations, both arrays must have the same shape or a single number can be used.<br>
    NumPy provides functions that are equivalent to the operators menthioned above.
</div>

```python
arr1 = np.random.randint(10, 60, size=(3, 4))
arr2 = np.random.randint(1, 10, size=(3, 4))
```

```python
arr1, arr2
```

```python
# same as arr1 + arr2
np.add(arr1, arr2)
```

```python
# same as arr1 - arr2
np.subtract(arr1, arr2)
```

```python
# same as arr1 * arr2
np.multiply(arr1, arr2)
```

```python
# same as arr1 / arr2
np.divide(arr1, arr2)
```

```python
# same as arr1 // arr2
np.floor_divide(arr1, arr2)
```

```python
# same as arr1 % arr2
np.mod(arr1, arr2)
```

```python
# same as arr1 ** arr2
np.power(arr1, arr2)
```

```python
# a single value can be used instead of an array
random_number = np.random.randint(50)
print(random_number)
np.add(arr1, random_number)
```
